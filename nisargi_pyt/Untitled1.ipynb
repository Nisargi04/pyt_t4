```python
def air_pollution():
    api_key='b9de1d23cbaaa6467aa06710caf5a3bd'
    request_url=f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid='+api_key
    response=requests.get(request_url)
    air_pollution=response.json()
    print(json.dumps(air_pollution,indent=5))
    aqi=air_pollution['list'][0]['main']['aqi']
    print("aqi-index",aqi)
    aqi_dict={1:"good",2:"Fair",3:"Moderate",4:"Poor",5:"Very Poor"}
    print("quality",aqi_dict[aqi])
air_pollution()
```

```python
def weather_forecast():
    api_key='b9de1d23cbaaa6467aa06710caf5a3bd'
    city=input("enter city name")
    #request_url=f'http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}'
    request_url1='http://api.openweathermap.org/geo/1.0/direct?q='+city+'&appid='+api_key
    response1=requests.get(request_url1)
    coordinates=response1.json()
    lat=coordinates[0]['lat']
    lon=coordinates[0]['lon']
    print(lat,lon)
    request_url2=f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid='+api_key
    response2=requests.get(request_url2)
    five_day=response2.json()
    print(json.dumps(five_day,indent=5))
    data={"date_time":[],"temperature":[],"pressure":[],"humidity":[],"wind_speed":[],"visibility":[],"description":[]}
    for i in five_day['list']:
        data["date_time"].append(i['dt_txt'])
        data["temperature"].append(i['main']['temp'])
        data["pressure"].append(i['main']['pressure'])
        data["humidity"].append(i['main']['humidity'])
        data["wind_speed"].append(i['wind']['speed'])
        data["description"].append(i['weather'][0]['description'])
        data["visibility"].append(i['visibility'])
    print(data)
    import pandas as pd
    df=pd.DataFrame(data)
    print(df)
weather_forecast()
```
